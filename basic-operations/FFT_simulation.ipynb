```python
from scipy.fft import fft, fftfreq
import numpy as np

#Number of sample points
N = 600

#Sample spacing
T = 1/800
x = np.linspace(0.0,N*T, N, endpoint=False)
y = np.sin(50 * 2 * np.pi * x) + 0.5 * np.sin(80.0 * 2 * np.pi * x)
yf  = fft(y)


from scipy.signal.windows import blackman
w = blackman(N) #what is this filter blackman?
ywf = fft(y*w) 
xf = fftfreq(N,T)[:N//2] # what this line does?

import matplotlib.pyplot as plt
plt.semilogy(xf[1:N//2],  2/N *np.abs(yf[1:N//2]), '-b')
plt.semilogy(xf[1:N//2], 2 / N* np.abs(ywf[1:N//2]), '-r')
plt.legend(['FFT', 'FFT w. window'])
plt.grid()
plt.show()
```
